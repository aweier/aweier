```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#import scipy
#import pymannkendall as mk

# Data from water quality
data = {
    'Month': ['March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November'],
    'Daily Avg (mg/L)': [0.0602, 0.0344, 0.1613, 0.1639, 0.246, 0.0025, 0.0451, 0.0351, 0.1231],
}
# Create a Pandas DataFrame from the data
df = pd.DataFrame(data)
print(df)
```

```python
# Plotting the data
#plt.rcParams['axes.facecolor'] = '#ED7D31'  # Orange color for axes background
#plt.rcParams['figure.facecolor'] = '#69b3a2'  # Teal color for figure background
#plt.rcParams['text.color'] = 'white'  # White color for text
plt.style.use('classic')
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Daily Avg (mg/L)'], marker='o', linestyle='-', color='#228B22')

# Adding titles and labels
plt.title('Monthly Daily Average Water Quality')
plt.xlabel('Month')
plt.ylabel('Daily Avg (mg/L)')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45)

# Adding grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
