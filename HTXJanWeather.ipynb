```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
degree_symbol = '\u00b0'
```

```python
df = pd.read_csv('htemp.csv')
df.head()


ax = sns.barplot(x='Date', y='Avg', data=df, hue='Avg', legend=False, palette='dark:#69b3a2')
# Remove x-axis ticks
plt.gca().set_xticks([1, 7, 14, 21, 28])
plt.ylabel(f'Temperature {degree_symbol}F')
plt.xlabel('Day of Month')
#Title
plt.title("Average Temperatures of Houston in January")
#Adjust ylim, xlim
lowest_temp_row = df.loc[df['Min'].idxmin()]
plt.ylim(df['Avg'].min() - 10, df['Avg'].max() + 10)
ax.annotate(f' {lowest_temp_row["Min"]} {degree_symbol}F', (lowest_temp_row.name, lowest_temp_row["Min"]), 
            xytext=(0, 5), textcoords='offset points', ha='center', rotation=90)
```
